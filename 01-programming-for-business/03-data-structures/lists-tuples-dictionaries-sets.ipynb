# List comprehension

```python
list1=[1,2,3,4]
list2=[]
```

```python
for x in list1:
    list2.append(x**2)
    
print(list2)
```

```python
list1=[1,2,3,4]
list2=[x for x in list1]
```

```python
list2
```

```python
list3=[1,12,2,20,3,15,4]
list4=[x for x in list1 if x <= 15]

list4
```

### Local Variables

A local variable is created inside a function and cannot be accessed by statements that are outside the function. Different functions can have local variables with the same names because the functions cannot see each other's local variables.


Anytime you assign a value to a variable inside a function, you create a local variable. A local variable belongs to the function in which it is created, and only statements inside that function can access the variable. (The term local is meant to indicate that the variable can be used only locally, within the function in which it is created.)

## Global Variables and Global Constants

# Global Variable

A global variable is accessible to all the functions in a program file. Global variables are variables that are declared outside of any function and can be accessed from any part of the program.

1. They are available throughout the entire program.
2. Can be modified inside a function using the global keyword.
3. They should be used cautiously as excessive use of global variables can make debugging and maintenance difficult.

```python
# Global variable
count = 10

def increment_count():
    global count  # Accessing and modifying the global variable
    count += 1
    print(f"Count inside function: {count}")

increment_count()  # Calling function to increment count
print(f"Count outside function: {count}")
```

# Global Constants

Global constants are similar to global variables but are intended to be immutable (unchangeable). In Python, there is no strict mechanism for defining constants, but by convention, variable names written in ALL CAPS are treated as constants.

1. They are usually declared at the start of the program.
2. Intended to remain unchanged throughout the program.
3. While you can technically modify them, it is discouraged.

```python
# Global constant
PI = 3.14159

def calculate_circumference(radius):
    print(2 * PI * radius)  # Using the global constant
```

```python
circumference = calculate_circumference(5)
```

```python
# Create a global variable.
my_value = 10

# The show_value function prints
# the value of the global variable.
def show_value():
    print(my_value)
```

```python
# Call the show_value function.
show_value()
```

```python
# The following is used as a global constant to represent
# the contribution rate.
CONTRIBUTION_RATE = 0.05
def main():
    gross_pay = float(input('Enter the gross pay: '))
    bonus = float(input('Enter the amount of bonuses: '))
    show_pay_contrib(gross_pay)
    show_bonus_contrib(bonus)

#The show_pay_contrib function accepts the gross
# pay as an argument and displays the retirement
# contribution for that amount of pay.
 
def show_pay_contrib(gross):
    contrib = gross * CONTRIBUTION_RATE
    print(f'Contribution for gross pay: ${contrib:,.2f}.')

# The show_bonus_contrib function accepts the
# bonus amount as an argument and displays the
# retirement contribution for that amount of pay.
def show_bonus_contrib(bonus):
    contrib = bonus * CONTRIBUTION_RATE
    print(f'Contribution for bonuses: ${contrib:,.2f}.')
```

```python
# Call the main function.
main()
```

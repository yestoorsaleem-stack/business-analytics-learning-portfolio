```python
import pandas as pd
data = pd.read_csv(r'C:\Users\Bilal Qaisar\Desktop\span_new.csv')
print(data.head())
```

```python
import matplotlib.pyplot as plt
```

# Line Plot: Months Old vs Price

```python
plt.plot(data['months_old'], data['price'])
plt.title('Price vs Months Old')
plt.xlabel('Months Old')
plt.ylabel('Price')
plt.show()
```

```python
plt.scatter(data['kms'], data['power'])
plt.title('Scatter Plot Example')
plt.xlabel('kms')
plt.ylabel('power')
plt.show()
```

Bar Plot: Illustrate categorical comparisons

# Bar plot: Average price for each fuel type

```python

avg_price_by_fuel = data.groupby('fuel_type')['price'].mean()
plt.bar(avg_price_by_fuel.index, avg_price_by_fuel.values)
plt.title('Average Price by Fuel Type')
plt.xlabel('Fuel Type')
plt.ylabel('Average Price')
plt.show()
```

# Histogram: Price distribution

```python

plt.hist(data['price'], bins=20, edgecolor='black')
plt.title('Price Distribution')
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.show()
```

# Scatter plot: Kms vs Price

```python

plt.scatter(data['kms'], data['price'], alpha=0.6)
plt.title('Kms vs Price')
plt.xlabel('Kms Driven')
plt.ylabel('Price')
plt.show()
```

# Box plot: Price distribution by fuel type

```python
data.boxplot(column='price')
plt.title('Price Distribution by Fuel Type')
plt.xlabel('Fuel Type')
plt.ylabel('Price')
plt.show()
```

# Line and Scatter plot: Months Old vs Price

```python
plt.plot(data['months_old'], data['price'], label='Line Plot')
plt.scatter(data['months_old'], data['price'], color='red', label='Data Points', alpha=0.5)
plt.title('Price vs Months Old')
plt.xlabel('Months Old')
plt.ylabel('Price')
plt.show()
```

# Pie chart: Distribution of fuel type

```python

fuel_counts = data['fuel_type'].value_counts()
plt.pie(fuel_counts, labels=fuel_counts.index, startangle=90)
plt.title('Fuel Type Distribution')
plt.show()
```

# Subplots: Kms vs Price and Months Old vs Price

```python
plt.figure(figsize=(8,6))

# Scatter plot: Kms vs Price
plt.subplot(1, 2, 1)
plt.scatter(data['kms'], data['price'], color='blue')
plt.xlabel('Kms')
plt.ylabel('Price')
plt.title('Kms vs Price')

# Line plot: Months Old vs Price
plt.subplot(1, 2, 2)
plt.plot(data['months_old'], data['price'], color='green')
plt.xlabel('Months Old')
plt.ylabel('Price')
plt.title('Months Old vs Price')

plt.show()
```

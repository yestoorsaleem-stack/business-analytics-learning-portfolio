```python
missing_price = data.loc[data['price'].isnull()]
print(missing_price)
```

```python
# Backward fill (bfill) for missing values in the 'price' column
data['price_bfill'] = data['price'].fillna(method='bfill')

# Forward fill (ffill) for missing values in the 'price' column
data['price_ffill'] = data['price'].fillna(method='ffill')

# Display the updated DataFrame
print(data[['price', 'price_bfill', 'price_ffill']])
```

* Backward Fill (bfill):

Replaces missing values with the next non-missing value in the column.
For example, if the price value in row 2 is missing, it will take the value from row 3.

* Forward Fill (ffill):

Replaces missing values with the previous non-missing value in the column.
For example, if the price value in row 2 is missing, it will take the value from row 1.

```python
# Calculate mean, median, and mode of the 'price' column
price_mean = data['price'].mean()
data['price_filled_mean'] = data['price'].fillna(price_mean)
price_median = data['price'].median()
price_mode = data['price'].mode()[0]  # Mode returns a Series, take the first value

# Fill missing values using mean
data['price_filled_mean'] = data['price'].fillna(price_mean)

# Fill missing values using median
data['price_filled_median'] = data['price'].fillna(price_median)

# Fill missing values using mode
data['price_filled_mode'] = data['price'].fillna(price_mode)

# Display the updated DataFrame
print(data[['price', 'price_filled_mean', 'price_filled_median', 'price_filled_mode']])
```

```python
# Check missing values
print(data[['make', 'model']].isnull().sum())
```

```python
# Drop rows where 'make' or 'model' is missing
data_dropped = data.dropna(subset=['make', 'model'])
print(data_dropped)
```

# Method 2Fill Missing Values with a Placeholder

```python
# Fill missing values with a placeholder (e.g., 'Unknown')
data['make_filled'] = data['make'].fillna('Unknown')
data['model_filled'] = data['model'].fillna('Unknown')
print(data[['make', 'model', 'make_filled', 'model_filled']])
```

# Method 3 Fill with the Most Frequent Value

```python
# Fill missing values with the most frequent value (mode)
make_mode = data['make'].mode()[0]
model_mode = data['model'].mode()[0]

data['make_mode'] = data['make'].fillna(make_mode)
data['model_mode'] = data['model'].fillna(model_mode)
print(data[['make', 'model', 'make_mode', 'model_mode']])
```

# Method 4 # Fill missing values with the previous value

```python
data['make_ffill'] = data['make'].fillna(method='ffill')
data['model_ffill'] = data['model'].fillna(method='ffill')
print(data[['make', 'model', 'make_ffill', 'model_ffill']])
```

# Method 5 Fill missing values with the next value

```python

data['make_bfill'] = data['make'].fillna(method='bfill')
data['model_bfill'] = data['model'].fillna(method='bfill')
print(data[['make', 'model', 'make_bfill', 'model_bfill']])
```

# sorting and Filtering

```python
sorted_data = data.sort_values(by='price', ascending=False)  # Sort by price
gasoline_cars = data[data['fuel_type'] == 'gasoline']  # Filter gasoline cars
expensive_cars = data[data['price'] > 500000]  # Filter expensive cars
filtered_data = data.query('price > 500000 & kms < 50000')  # Filter with conditions
print(filtered_data)
```

# Aggregation

```python
fuel_avg_price = data.groupby('fuel_type')['price'].mean()  # Group by fuel type
print(fuel_avg_price)

agg_stats = data.groupby(['make', 'fuel_type'])['price'].agg(['mean', 'max', 'min'])  # Aggregation
print(agg_stats)
```

# If statement


The __if__ statement_ is used to create a decision structure, which allows a program to have more than one path of execution. 

The __if__ statement causes one or more statements to execute only when a Boolean expression is true.

A __control structure__ is a logical design that controls the order in which a set of statements 
execute. 

A __sequence structure__ is a set of statements that execute in the order in which they appear.

In a __decision structure’s__ simplest form, a specific action is performed only if a certain condition exists.

```python
base_hours = 40
Overtime_multiplier = 1.5

hours_worked = float(input("enter the total number of hours worked "))
pay_rate = float(input("enter the hourly pay rate "))

if hours_worked > base_hours:
    overtime_hours_worked = hours_worked - base_hours    ## Calculate the overtime hours worked
    
    overtime_pay = overtime_hours_worked * pay_rate * Overtime_multiplier  
    
    Gross_pay = (base_hours * pay_rate) + overtime_pay 
    
else:
    Gross_pay = hours_worked * pay_rate
          
    

print(f'Gross pay is {Gross_pay}')
    
    
    
```

## Nested If 

```python
## Class practice 

inter_marks = int(input("Enter your Inter marks "))
entry_test_score = int(input("enter your entry test score "))

if inter_marks >= 80:
    if entry_test_score >= 75:
        print("you are eligible for the admission ")
    else:
        print("you have low entry test score ")
else:
    print("sorry! you are not eligible for the admission ")
```

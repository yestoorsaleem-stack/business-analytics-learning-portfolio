# 1.	Grade Calculator

```python
score=int(input("Enter your score out of 100 "))

if score>=90:
    print ("You got A grade")
elif score>=80 and score<=89:
    print ("You got B grade")
elif score>=70 and score<=79:
    print ("You got C grade")
elif score>=60 and score<=69:
    print ("You got D grade")
elif score<60:
    print ("You got F grade")
else:
    print ("Invalid input")
```

# 3.	Number Sign Checker

```python
number= int(input("Enter a number"))
if number>0:
    print("Entered number is positive")
elif number<0:
    print("Entered number is negitive")
elif number==0:
    print ("Entered number is zero")
```

# 4.	Simple Calculator

```python
number1=int(input("Enter first number "))
number2=int(input("Enter second number "))
operator=input("Enter operator (+ , - ,*, /)")

if operator=="+":
    result=number1+number2
    print ("Addition of two numbers is ", result)
elif operator=="-":
    result=number1-number2
    print ("subtraction of two numbers is ", result)
elif operator=="*":
    result=number1*number2
    print ("Multiplication of two numbers is ", result)
elif operator=="/":
    if number2!=0:
        result =number1/number2
        print ("Division of two numbers  is ", result)
    else:
        print ("Division by zero is not allowed")
else:
    print ("Invalid input")
```

# 8.	Discount Calculator

```python
purchase_amount = float(input("Enter the total purchase amount: "))

if purchase_amount > 500:
    discount_percentage = 0.20
elif purchase_amount > 100:
    discount_percentage = 0.10


discount_amount = discount_percentage * purchase_amount
final_amount = purchase_amount - discount_amount

print(purchase_amount)
print(discount_percentage)
print(final_amount)
```

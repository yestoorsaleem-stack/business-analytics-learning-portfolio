## While Loop

1- Declare an initialization value first 
 
2- Write the while condition
 
3- Set of statements to execute
 
4- End the program with resetting of initialization value OR an input statement which asks the user about the continuation of the loop .

```python
num=1
while num<=10:
    print(num)
    num+=1
```

```python
num=10
while num>=1:
    print(num)
    num=num-1
```

```python
num=1
print(f'number\t\tsqaure')
while num<=10:
    print(f'{num}\t\t{num**2}')
    num+=1
```

```python
num=1
total=0
while num>0:
    num=int(input("Enter positive number"))
    if num>0:
        total=total+num
    
print(total)
```

## Nested Loop: A loop that is inside another loop is called a nested loop.


```python
for x in range(4):
    for y in range(2):
        print(f'for each value of x:{x}, y = {y}')
```

```python
for i in range(3):
    for j in range(3):
        print("*",end="")
    print("\n")
```

```python
for row in range(8):
    for col in range(6):
        print('#', end='')
    print()
```

```python
num=1
for i in range(3):
    for j in range(3):
        print(num,end=" ")
        num=num+1
    print('\n')
```

### Displaying Formatted Output with F-strings

 An __f-string__ can contain placeholders for variables and other valid expressions.

```python
print(f'Hello world')

temperature = 45
print(f'It is currently {temperature * 2} degrees.')

print(f'The value is {10 + 2}.')

val = 10
print(f'The value is {val + 2}.')
```

```python
val=20.15875
print (f'the value is, {val:.2f}')
```

```python
amount = 12.9876
print(f'Amount is {amount:.2f}')

marks = 0.56
print(f'Percentage = {marks:%}')

print(f'Percentage = {marks:.0%}')
```

# Comparison Operators

These operators compare two values and return a boolean result (True or False).

* == : Equal to (e.g., 5 == 5 returns True)
* != : Not equal to (e.g., 5 != 3 returns True)
* "> : Greater than (e.g., 5 > 3 returns True)"
* "< : Less than (e.g., 5 < 3 returns False)"
* ">= : Greater than or equal to (e.g., 5 >= 5 returns True)"
* "<= : Less than or equal to (e.g., 5 <= 3 returns False)"



## Logical Operators


These operators are used to combine conditional statements.

* and : Returns True if both statements are true (e.g., True and False returns False)
* or : Returns True if at least one statement is true (e.g., True or False returns True)
* not : Reverses the result, returns False if the statement is true (e.g., not True returns False)

```python
True and True #Logical AND: True if all conditions are True
```

```python
(5 > 2) and (7 > 2) #True if all conditions are True
```

```python
(5 > 2) & (7 > 2) #True if all conditions are True
```

## Assignment Operators


These operators assign values to variables.

* "= : Assigns value (e.g., x = 5)
* "+= : Adds and assigns (e.g., x += 3 is equivalent to x = x + 3)
* "-= : Subtracts and assigns (e.g., x -= 3)
* "*= : Multiplies and assigns (e.g., x *= 3)
